```python
# Creating a fuction to check null valus present in column
def null_values(df1):
    return round((df1.isnull().sum()*100/len(df1)).sort_values(ascending=False),2)
```

# Bining

Create the bins for entity  AMT_INCOME_TOTAL and AMT_CREDIT to identify leads

```python
bins = [0,25000,50000,75000,100000,125000,150000,175000,200000,225000,250000,275000,300000,325000,350000,375000,400000,10000000]
income_slot = ['0-25k', '25k-50k','50k-75k','75k,1lac','1lac-1.25lac', '1.25lac-1.50lac', '1.5lac-1.75lac','1.75lac-2lac',
       '2lac-2.25lac','2.25lac-2.5lac','2.5lac-2.75lac','2.75lac-3lac','3lac-3.25lac','3.25lac-3.5lac','3.5lac-3.75lac',
       '3.75-4lac','4 lac and above']

df1['AMT_INCOME_RANGE']=pd.cut(df1['AMT_INCOME_TOTAL'],bins,labels=income_slot)
```

```python
binss = [0,25000,50000,75000,100000,125000,150000,175000,200000,225000,250000,275000,300000,325000,350000,375000,400000,10000000]
credit_slots = ['0-25k', '25k-50k','50k-75k','75k,1lac','1lac-1.25lac', '1.25lac-1.50lac', '1.5lac-1.75lac','1.75lac-2lac',
       '2lac-2.25lac','2.25lac-2.5lac','2.5lac-2.75lac','2.75lac-3lac','3lac-3.25lac','3.25lac-3.5lac','3.5lac-3.75lac',
       '3.75-4lac','4 lac and above']

df1['AMT_CREDIT_RANGE']=pd.cut(df1['AMT_CREDIT'],bins=binss,labels=credit_slots)
```
